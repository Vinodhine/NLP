```python
import string
with open('Obamacare_Red_Consolidated.csv','r') as infile:
    content = infile.read()
    st = content.splitlines()
    s=" ".join(str(x) for x in st)
#print s
```

```python
new_lst=[]
p = string.punctuation
d = string.digits
print(p)
print(d)
```

```python
No_punct_lst=[]
No_digits_lst = []
table_p = string.maketrans(p, len(p) * " ")
table_d = string.maketrans(d, len(d) * " ")
#print(type(table_p))
#print(type(new_lst))
#print(s.translate(table_p))
#print(s.translate(table_p))
#print(s.translate(table_d))
No_punct_lst.append(s.translate(table_p))
str_without_punct = ''.join(No_punct_lst)
#print str_without_punct
No_digits_lst.append(str_without_punct.lower().translate(table_d))
#print No_digits_lst
```

```python
#print No_digits_lst
text_lowercase= ''.join(No_digits_lst)
#print type(text_lowercase)
#print len(text_lowercase)
text_to_wordlist = text_lowercase.split()
#print text_to_wordlist
```

```python
import pickle
with open('Obamacare_consolidated_Red.txt', 'wb') as fp:
    pickle.dump(No_digits_lst, fp)
```
