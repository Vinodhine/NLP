```python
with open('My_tweets_trumpcare.json', 'r') as f:#reads in the JSON file into Python as a string
    t = json.load(f)
print len(t)
```

```python
csv_out = open('My_tweets_Trump.csv','a') #opens csv file
writer = csv.writer(csv_out) #create the csv writer object
fields = ['text','followers','screen_name','location'] #field names
writer.writerow(fields) 
#writer.writerow(fields) #writes field
```

```python
for line in t:
 
    #writes a row and gets the fields from the json object
    #screen_name and followers/friends are found on the second level hence two get methods
    writer.writerow([line.get('text').encode('unicode_escape'),
                     line.get('user').get('screen_name'),
                     line.get('user').get('followers_count'),
                     line.get('location')])
                     #line.get('location')])#.encode('utf-8')])
                     #line.get('location')])
    #writer.writerow([line.get('text').encode('unicode_escape')])
 

csv_out.close()
```
