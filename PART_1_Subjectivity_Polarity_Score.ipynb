```python
from textblob import TextBlob
with open('Obamacare_Red_Consolidated.csv','r') as infile:
    content = infile.read()
    sentences = content.splitlines()
#print sentences
#print len(sentences)
print len(sentences)
print sentences[0]
print sentences[-1]
```

```python
sub_list = []
pol_list = []
for s in sentences:
    s1 = s.decode('utf-8')
    tb = TextBlob(s1)
    sub_list.append(tb.sentiment.subjectivity)
    pol_list.append(tb.sentiment.polarity)
print (sum(sub_list)/len(sub_list))
print (sum(pol_list)/len(pol_list))
```

```python
import matplotlib.pyplot as plt

plt.hist(sub_list, bins=15) #, normed=1, alpha=0.75)
plt.xlabel('Subjectivity score')
plt.ylabel('Sentence count')
plt.grid(True)
plt.title('Subjectivity Score')
plt.savefig('subjectivity_Obamacare-Red.pdf')
plt.show()
```

```python
plt.hist(pol_list, bins=20) #, normed=1, alpha=0.75)
plt.xlabel('Polarity Score')
plt.ylabel('Sentence count')
plt.grid(True)
plt.title('Polarity Score')
plt.savefig('Polarity_Obamacare-Red.pdf')
plt.show()
```
