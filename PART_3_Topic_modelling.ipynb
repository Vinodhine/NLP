```python
import string
p = string.punctuation
d = string.digits
#print(p)
#print(d)
with open('Obamacare_Red_Consolidated.csv','r') as infile:
    content = infile.read()
    list1 = content.lower().splitlines()
#list1 = ['life is good##3 sh et','https:\\Hell a lot of problems ate','7537Https@ Trump is great ']
no_punctuation = []
no_digits = []
for i in list1:
    #print i
    table_p = string.maketrans(p, len(p) * " ")
    table_d = string.maketrans(d, len(d) * " ")
    #print table_p
    #print table_d
    no_punctuation.append(i.translate(table_p))
    #no_digits.append(str_without_punct.lower().translate(table_d))
    #print no_punctuation
    #print no_digits
    
print len(no_punctuation)

for i in no_punctuation:
    no_digits.append(i.lower().translate(table_d))
print len(no_digits)
```

```python
#list2 = []
import nltk
stopwords = nltk.corpus.stopwords.words('english')
#print type(stopwords), len(stopwords), stopwords

lst2=['trump','donald','http','https','u','n','via' , 'btw' , 'hey', 'amp','say',
      'president','nhttps','russia','obama','would','like','one','new','news']
stopwords=stopwords+lst2
#print stopwords



list_of_str = []
for i in no_digits:
    str = []
    str = i.split()
    #print str
    list2 = []
    #list_without_stopwords2 = []
    for word in str:
        if word not in stopwords and len(word)>3:
            #list2 = []
            list2.append(word)
            
           
    str_without_punct = ' '.join(list2)
    #print str_without_punct
    list_of_str.append(str_without_punct)
print list_of_str[:11]
```

```python
outfile = open('Final_corpus_for_topic_modelling_Obama_RedStates.txt', 'w')
for tweet in list_of_str:
    outfile.write(tweet)
    outfile.write('\n')
    
outfile.close()
```
